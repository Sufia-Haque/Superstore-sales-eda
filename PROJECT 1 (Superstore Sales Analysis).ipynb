```python
df = pd.read_csv(r'C:\Users\lenovo\Desktop\sufia\Superstore dataset.csv', encoding = 'latin-1')
print("Dataset loaded!")
print(f"Shape: {df.shape[0]} rows and {df.shape[1]} columns")
```

```python
# Check column names and data types
df.info()
```

```python
# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())
```

```python
# Basic statistics
df.describe()
```

```python
# ============================================
# STEP 4: Clean the Data
# ============================================

# Convert Order Date and Ship Date to proper date format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Extract year and month for time analysis
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month

print("✅ Data cleaned and dates formatted!")
print(df[['Order Date', 'Ship Date', 'Year', 'Month']].head())
```

```python
# ============================================
# Q1: Sales and Profit by Category
# ============================================

category_summary = df.groupby('Category')[['Sales', 'Profit']].sum().reset_index()
print(category_summary)

# Plot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Sales chart
axes[0].bar(category_summary['Category'], category_summary['Sales'], color=['#4C72B0','#DD8452','#55A868'])
axes[0].set_title('Total Sales by Category', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Sales ($)')

# Profit chart
axes[1].bar(category_summary['Category'], category_summary['Profit'], color=['#4C72B0','#DD8452','#55A868'])
axes[1].set_title('Total Profit by Category', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Category')
axes[1].set_ylabel('Profit ($)')

plt.tight_layout()
plt.savefig('chart1_category.png', dpi=150)  # saves chart for GitHub README
plt.show()
```

```python
# ============================================
# Q2: Sales and Profit by Region
# ============================================

region_summary = df.groupby('Region')[['Sales', 'Profit']].sum().reset_index()
region_summary['Profit Margin %'] = (region_summary['Profit'] / region_summary['Sales'] * 100).round(2)
print(region_summary)

sns.barplot(data=region_summary, x='Region', y='Profit Margin %', palette='Blues_d')
plt.title('Profit Margin % by Region', fontsize=14, fontweight='bold')
plt.ylabel('Profit Margin (%)')
plt.savefig('chart2_region.png', dpi=150)
plt.show()
```

```python
# ============================================
# Q3: Customer Segment Analysis
# ============================================

segment = df.groupby('Segment')[['Sales', 'Profit']].sum().reset_index()

fig, ax = plt.subplots()
ax.pie(segment['Sales'], labels=segment['Segment'], autopct='%1.1f%%',
       colors=['#4C72B0','#DD8452','#55A868'], startangle=90)
ax.set_title('Sales Share by Customer Segment', fontsize=14, fontweight='bold')
plt.savefig('chart3_segment.png', dpi=150)
plt.show()
```

```python
# ============================================
# Q4: Sales Trend Over Time
# ============================================

monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()
monthly_sales['Date'] = pd.to_datetime(monthly_sales[['Year', 'Month']].assign(day=1))

plt.plot(monthly_sales['Date'], monthly_sales['Sales'], marker='o', color='#4C72B0', linewidth=2)
plt.title('Monthly Sales Trend Over Time', fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('chart4_trend.png', dpi=150)
plt.show()
```

```python
# ============================================
# Q5: Top 10 Products Losing Money
# ============================================

loss_products = df[df['Profit'] < 0].groupby('Sub-Category')['Profit'].sum().sort_values().head(10).reset_index()

sns.barplot(data=loss_products, x='Profit', y='Sub-Category', palette='Reds_r')
plt.title('Sub-Categories with Biggest Losses', fontsize=14, fontweight='bold')
plt.xlabel('Total Profit ($)')
plt.tight_layout()
plt.savefig('chart5_losses.png', dpi=150)
plt.show()
```
